```python
from math import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import csv
import time
```

```python
ndt_position = np.genfromtxt('ndt_position.csv',delimiter=',')
xsens_position = np.genfromtxt('xsens_position.csv',delimiter=',')
projected_heading = np.genfromtxt('lane.csv', delimiter=',')
white_line = np.genfromtxt('whiteline.csv', delimiter=',')
#imu_data = np.genfromtxt('imu.csv', delimiter=',')
```

```python
toDelete = []
for idx, i in enumerate(white_line):
    if i[0] == 0 or i[1] == 0 or i[3] == 0 or i[4] == 0:
        toDelete.append(idx)
white_line = np.delete(white_line, toDelete, 0)
```

```python
fig = plt.figure(figsize=(32, 18))
ax = fig.add_subplot(1, 1, 1)
plt.title('Fusion Results', fontsize = 40)
plt.xlabel('East', fontsize = 40)
plt.ylabel('North', fontsize = 40)

plt.scatter(ndt_position[:, 0], ndt_position[:, 1], c = 'gray')
#plt.plot(ndt_position[:, 0], ndt_position[:, 1], c = 'gray')

plt.scatter(xsens_position[:, 0], xsens_position[:, 1], c = 'green')
#plt.plot(xsens_position[:, 0], xsens_position[:, 1], c = 'green')

#plt.scatter(imu_data[:, 0], imu_data[:, 1], c = 'magenta')

plt.scatter(projected_heading[:, 0], projected_heading[:, 1], c = 'orange')
plt.plot(projected_heading[:, 0], projected_heading[:, 1], c = 'orange')

plt.scatter(projected_heading[0, 0], projected_heading[0, 1], c = 'black', s = 100)

s = 160
plt.scatter(ndt_position[s, 0], ndt_position[s, 1], c = 'magenta', s = 50)
plt.scatter(xsens_position[s, 0], xsens_position[s, 1], c = 'black', s = 50)
plt.scatter(projected_heading[s, 0], projected_heading[s, 1], c = 'cyan', s = 50)


for line_idx, line in enumerate(white_line):
    if line_idx % 2 == 0:
        plt.plot([line[0], line[3]], [line[1], line[4]], color = 'red', linewidth = 5)
    else:
        plt.plot([line[0], line[3]], [line[1], line[4]], color = 'blue', linewidth = 5)

x_ticks = range(6720, 6860, 20)
ax.set_xticklabels(x_ticks, rotation = 0, fontsize = 40)
y_ticks = range(-8340, -8160, 20)
ax.set_yticklabels(y_ticks, rotation = 0, fontsize = 40)
```
